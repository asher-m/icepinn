```python
import matplotlib.pyplot as plt
import numpy as np
from sympy import *


x, r, s, b = symbols('x, r, s, b', real=True)
```

```python
u = 2 * x / r
B3 = (
    2 / r * Piecewise(
        ((4 - 6 * u ** 2 + 3 * Abs(u) ** 3) / 6, Abs(u) < 1),
        (((2 - Abs(u)) ** 3) / 6, Abs(u) < 2),
        (0, True),
    )
).subs(r, 1)
B3 = B3.simplify()

# for demonstration purposes
t = np.linspace(-2, 2, 1000)
plt.plot(
    t,
    lambdify([x], B3)(t)
)
plt.show()

print(
    B3
)
```

```python
I1 = integrate(
    -8*(Abs(x) - 1)**3/3,
    (x, -1, b)
).refine(And(b >= -1, b <= -0.5))
print(
    pycode(
        nsimplify(
            I1,
            rational=True
        ),
    ),
    nsimplify(
        I1.subs(b, -0.5),
        rational=True
    ),
    sep='\n'
)
```

```python
I2 = integrate(
    8*x**2*Abs(x) - 8*x**2 + 4/3,
    (x, -0.5, b)
).refine(And(b >= -0.5, b <= 0.5))
print(
    pycode(
        nsimplify(
            I2,
            rational=True
        ),
    ),
    nsimplify(
        I2.subs(b, 0.5),
        rational=True
    ),
    sep='\n'
)
```
